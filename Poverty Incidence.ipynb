```python
import pandas as pd
import geopandas as gpd
```

```python
poverty_incidence_df = pd.read_csv("dataset/Poverty_Incidence/Povery_Incidence_OM_2009_2012_2015.csv")
```

```python
poverty_incidence_df['NAME_2'] = poverty_incidence_df['municipality']
poverty_incidence_df
```

```python
ph_cities_df = gpd.read_file('PH_ShapeFile/gadm41_PHL_shp/gadm41_PHL_2.shp')
ph_cities_df.plot(figsize=(20,10))
```

```python
ph_cities_df.head()
```

```python
om_cities_df = ph_cities_df[ph_cities_df['NAME_1'] == 'Oriental Mindoro']
# om_cities_df = om_cities_df[['COUNTRY', 'NAME_2', 'geometry']]
om_cities_df.head()
```

```python
om_cities_df.plot()
```

```python
merged_cities_with_poverty = om_cities_df.merge(poverty_incidence_df, on='NAME_2')
merged_cities_with_poverty = merged_cities_with_poverty.drop(columns=['municipality'])
merged_cities_with_poverty
# cities_with_poverty = gpd.sjoin(poverty_incidence_df, ph_cities_df, how='right', op='intersects')
```

```python
import plotly.express as px

fig = px.choropleth(merged_cities_with_poverty, 
                    geojson=merged_cities_with_poverty.geometry,
                    locations=merged_cities_with_poverty.index,
                    color='pi_2009',
                    # scope='district',
                    color_continuous_scale='Viridis')

fig.show()
```
